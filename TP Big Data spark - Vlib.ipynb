```
df = spark.read.format("csv").option("inferSchema", "true").option("header", "true").option("sep", ',').load("dbfs:/FileStore/shared_uploads/nedjemeddine.daoud@etu.univ-amu.fr/201801_fordgobike_tripdata.csv")
```

```
df.display()
```

```
df.printSchema()
```

```
# dropping un-neccessary columns
df.drop("start_station_id","end_station_id").printSchema()
```

```
# missing value in each columns
from pyspark.sql.functions import col,sum
df.select(*(sum(col(c).isNull().cast("int")).alias(c) for c in df.columns)).show()
```

```
# remove milliseconds
from pyspark.sql.functions import to_date

from pyspark.sql.functions import substring, length, col, expr
df = df.withColumn("start_time",expr("substring(start_time, 1, length(start_time)-5)"))
df = df.withColumn("end_time",expr("substring(end_time, 1, length(end_time)-5)"))
display(df)
```

```
# changing datatype of 'start_time' and 'end_time' columns to datetime
from pyspark.sql.functions import to_timestamp
df = df.withColumn('start_time',to_timestamp(df.start_time, 'yyyy-MM-dd HH:mm:ss'))
df = df.withColumn('end_time',to_timestamp(df.end_time, 'yyyy-MM-dd HH:mm:ss'))
```

```
# separating start_time_date, start_time_hour, start_time_day, start_time_month from start_time column
from pyspark.sql.functions import col,hour,next_day,date_format,month,to_date
df = df.withColumn('start_time_date',to_date(df.start_time))
df = df.withColumn('start_time_hour',hour(df.start_time))
df = df.withColumn('start_time_day', date_format('start_time', 'EEEE'))
df.printSchema()
```

```
# separating end_date, end_hour, end_day, end_month from end_time column
from pyspark.sql.functions import col,hour,next_day,date_format,month,to_date

df = df.withColumn('end_time_date',to_date(df.end_time))
df = df.withColumn('end_time_hour',hour(df.end_time))
df = df.withColumn('end_time_day', date_format('end_time', 'EEEE'))
df.printSchema()
```

```
# verifying date, hour, day and month are separated
df.display()
```

```
# creating new column contains ride duration in minute
df = df.withColumn('duration_min',df.duration_sec/60)
```

```
# Bike Rides by User Type
from pyspark.sql.functions import col

total = df.count()

display(df.groupBy("user_type") \
  .count() \
  .withColumnRenamed('count', 'cnt_per_group') \
  .withColumn('pourcentage', (col('cnt_per_group') / total) * 100 )
  .drop('cnt_per_group')
       )
```

```
# Bike Rides by Bike Share For All
from pyspark.sql.functions import col

total = df.count()

display(df.groupBy("bike_share_for_all_trip") \
  .count() \
  .withColumnRenamed('count', 'cnt_per_group') \
  .withColumn('pourcentage', (col('cnt_per_group') / total) * 100 )
  .drop('cnt_per_group')
       )
```

```
# Rides by Days of Week

from pyspark.sql.functions import col

total = df.count()

display(df.groupBy("start_time_day") \
  .count() \
  .withColumnRenamed('count', 'cnt_per_group') \
  .withColumn('pourcentage', (col('cnt_per_group') / total) * 100 )
  .drop('cnt_per_group')
       )
```

```
# Rides by Hours

from pyspark.sql.functions import col

total = df.count()

display(df.groupBy("start_time_hour") \
  .count() \
  .withColumnRenamed('count', 'cnt_per_group') \
  .withColumn('pourcentage', (col('cnt_per_group') / total) * 100 )
  .drop('cnt_per_group')
       )
```
